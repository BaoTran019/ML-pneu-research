```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../../data/File_10000_final.xlsx')
```

```python
df
```

```python
from sentence_transformers import SentenceTransformer

# Load model (dùng BERT hoặc một mô hình khác)
model = SentenceTransformer('FremyCompany/BioLORD-2023')

# Chuyển sang embedding
emb = model.encode(df['text'])
```

```python
emb_str = np.apply_along_axis(lambda row: ' '.join(map(str, row)), axis=1, arr=emb).reshape(-1, 1)
```

```python
df['BioLORD emb'] = emb_str
```
